## Make a plot using cartopy

```python
plt.figure(figsize=(10,5));
ax = plt.axes(projection=ccrs.PlateCarree());
ax.set_extent([30, 150, 10, 60])
ds['snow_fraction'].plot.pcolormesh(ax=ax, transform=ccrs.PlateCarree(),x='x', y='y', add_colorbar=False);
ax.coastlines();
plt.tight_layout();
```
